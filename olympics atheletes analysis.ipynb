```python
data[['Height','Weight','Year']].sort_values('Year',ascending=True).dropna(subset=['Height']).drop_duplicates(subset='Height')
```

```python
import seaborn as sns
```

```python
data1=data[['Height','Weight','Year']].sort_values('Year',ascending=True).dropna(subset=['Height']).drop_duplicates(subset='Height')
```

```python
plt.scatter(data1['Year'],data1['Height'],color='green',label='Height')
plt.scatter(data1['Year'],data1['Weight'],color='black',label='Weight')
plt.title('Height and Weight distribution over the years')
plt.xlabel('years')
plt.ylabel('height and weight')
plt.legend()
```

```python
medal_data = data[data['Medal'].notna()]
```

```python
medals_by_year = medal_data.groupby('Year').size()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(medals_by_year.index, medals_by_year.values, marker='o', linestyle='-', color='b')
plt.title('Number of Medals Won by Athletes Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.grid(True)
plt.show()
```
